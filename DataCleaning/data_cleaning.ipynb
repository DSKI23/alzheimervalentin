```python
from datasets import load_dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

alzheimer_df = load_dataset("DS23-KI-Projekt/alzheimerdataset")
df = alzheimer_df["train"].to_pandas()
df
```

```python
df.columns.isna()
```

```python
df.columns
```

```python
cleaned_df = df[['Age', 'BMI', 'Education Level', "Gender", 'Family History of Alzheimer’s', 'Cognitive Test Score',
       'Genetic Risk Factor (APOE-ε4 allele)', 'Alzheimer’s Diagnosis']]
cleaned_df
```

```python
cleaned_df.loc[:, 'Gender'] = cleaned_df.loc[:, 'Gender'].replace({'Male': 1, 'Female': 0})
cleaned_df.loc[:, 'Family History of Alzheimer’s'] = cleaned_df.loc[:, 'Family History of Alzheimer’s'].replace({'Yes': 1, 'No': 0})
cleaned_df.loc[:, 'Genetic Risk Factor (APOE-ε4 allele)'] = cleaned_df.loc[:, 'Genetic Risk Factor (APOE-ε4 allele)'].replace({'Yes': 1, 'No': 0})
cleaned_df.loc[:, 'Alzheimer’s Diagnosis'] = cleaned_df.loc[:, 'Alzheimer’s Diagnosis'].replace({'Yes': 1, 'No': 0})
cleaned_df
```
