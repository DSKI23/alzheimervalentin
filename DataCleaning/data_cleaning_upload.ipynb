```python
from datasets import load_dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

alzheimer_df = load_dataset("DS23-KI-Projekt/alzheimerdataset")
df = alzheimer_df["train"].to_pandas()
df
```

```python
df.columns.isna()
```

```python
df.columns
```

```python
cleaned_df = df[['Age', 'BMI', 'Education Level', "Gender", 'Family History of Alzheimer’s', 'Cognitive Test Score',
       'Genetic Risk Factor (APOE-ε4 allele)', 'Alzheimer’s Diagnosis']]
cleaned_df
```

```python
cleaned_df.loc[:, 'Gender'] = cleaned_df.loc[:, 'Gender'].replace({'Male': 1, 'Female': 0})
cleaned_df.loc[:, 'Family History of Alzheimer’s'] = cleaned_df.loc[:, 'Family History of Alzheimer’s'].replace({'Yes': 1, 'No': 0})
cleaned_df.loc[:, 'Genetic Risk Factor (APOE-ε4 allele)'] = cleaned_df.loc[:, 'Genetic Risk Factor (APOE-ε4 allele)'].replace({'Yes': 1, 'No': 0})
cleaned_df.loc[:, 'Alzheimer’s Diagnosis'] = cleaned_df.loc[:, 'Alzheimer’s Diagnosis'].replace({'Yes': 1, 'No': 0})
cleaned_df
```

Split und Upload des Datasets:

```python
from huggingface_hub import HfApi
from huggingface_hub import create_repo
from huggingface_hub import notebook_login
from datasets import load_dataset
from sklearn.model_selection import train_test_split
import pandas as pd
import os
```

```python
train_test_df, val_df = train_test_split(cleaned_df, test_size=0.2, random_state=42)
print(train_test_df.shape)
print(val_df.shape)
```

```python
os.mkdir('kaggle')

train_test_df.to_csv('/content/kaggle/training.csv', index=False)
val_df.to_csv('/content/kaggle/validation.csv', index=False)
```

```python
api = HfApi()
api.upload_folder(
    folder_path="/content/kaggle",
    repo_id="DS23-KI-Projekt/alzheimerdataset_split",
    repo_type="dataset"
)
```
