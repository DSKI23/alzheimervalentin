```python
from huggingface_hub import HfApi
from huggingface_hub import create_repo
from huggingface_hub import notebook_login
from datasets import load_dataset
from sklearn.model_selection import train_test_split
import pandas as pd
import os
```

```python
notebook_login()
```

```python
dataset = load_dataset("DS23-KI-Projekt/alzheimerdataset")
dataset
```

```python
create_repo("DS23-KI-Projekt/alzheimerdataset_split", repo_type="dataset")
'https://huggingface.co/datasets/DS23-KI-Projekt/alzheimerdataset_split'
```

```python
df = dataset['train'].to_pandas()
df.head()
```

```python
train_test_df, val_df = train_test_split(df, test_size=0.2, random_state=42)
print(train_test_df.shape)
print(val_df.shape)
```

```python
#os.mkdir('kaggle')

train_test_df.to_csv('/content/kaggle/train_test.csv', index=False)
val_df.to_csv('/content/kaggle/val.csv', index=False)
```

```python
api = HfApi()
api.upload_folder(
    folder_path="/content/kaggle",
    repo_id="DS23-KI-Projekt/alzheimerdataset_split",
    repo_type="dataset"
)
```
